```python
data.drop(columns=['type'], inplace=True)
```

```python
X = data.drop("quality",axis=1)
```

```python
y = data["quality"]
```

```python
X.columns
```

```python
from  sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
test_x
```

```python
train_x
```

```python
#pre processing (no missing no outlier etc) but we should perform scalling bec distance is involved
#y=mx+c bec scaling will smooth
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled_train = scaler.fit_transform(train_x)
```

```python
#SC -> no range is known (eg salary)
#minmax -> when range is known (eg 0 (min)-10 (max))
```

```python
#SVM
```

```python
import seaborn as sns
sns.pairplot(data,hue="quality")
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(scaled_train,train_y)
```

```python
#transform the data
scaled_test_x = scaler.transform(test_x)
```

```python
y_pred = model.predict(scaled_test_x)
```

```python
from sklearn.metrics  import accuracy_score
```

```python
round(accuracy_score(test_y,y_pred)*100,2)
```
